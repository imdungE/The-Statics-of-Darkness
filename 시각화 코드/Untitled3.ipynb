```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc("font", family=font)
plt.rcParams["axes.unicode_minus"] = False

# 1) CSV 파일 경로
private_path = r"data/사립일반고.csv"
public_path  = r"data/공립일반고.csv"

# 2) CSV 로드
df_private = pd.read_csv(private_path)
df_public  = pd.read_csv(public_path)

private_counts = df_private["키워드 등장 수"]
public_counts  = df_public["키워드 등장 수"]

# 3) 공통 구간(bin) 설정
max_count = max(private_counts.max(), public_counts.max())
bins = np.arange(0, max_count + 5, 5)  # 키워드 개수 단위로 구간 나눔

# 4) 도수분포표 (히스토그램) 생성
counts_pub, _ = np.histogram(public_counts, bins=bins)
counts_pri, _ = np.histogram(private_counts, bins=bins)

x = bins[:-1]
width = 2

plt.figure(figsize=(8, 5))

plt.bar(
    x - width/2,
    counts_pub,
    width=width,
    color="blue",
    edgecolor="blue",
    linewidth=1.5,
    alpha=0.7,
    label="공립일반고"
)

plt.bar(
    x + width/2,
    counts_pri,
    width=width,
    color="red",
    edgecolor="red",
    linewidth=1.5,
    alpha=0.7,
    label="사립일반고"
)

# 5) 그래프 설정
plt.xlabel("학교 주최 프로그램(키워드) 개수")
plt.ylabel("학생 수(도수)")
plt.title("일반고 유형별 학교 주최 프로그램 개수 도수분포 비교")
plt.legend()
plt.xticks(bins)
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터
pub = public_counts.values
pri = private_counts.values

y_pub = np.random.normal(1, 0.04, size=len(pub))
y_pri = np.random.normal(1.05, 0.04, size=len(pri))

plt.figure(figsize=(9, 4))

plt.scatter(
    pub, y_pub,
    color="blue",
    alpha=0.5,
    label="공립일반고",
    s=60
)

plt.scatter(
    pri, y_pri,
    color="red",
    alpha=0.5,
    label="사립일반고",
    s=60
)

plt.yticks([1.025], ["일반고"])
plt.xlabel("학교 주최 프로그램 키워드 개수")
plt.title("학교 유형별 학교 주최 프로그램 분포 (점 기반 표현)")
plt.grid(axis="x", linestyle="--", alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

pub = public_counts.values
pri = private_counts.values

plt.figure(figsize=(9, 3))

# 공립일반고 (파란 점)
plt.scatter(
    pub,
    np.zeros_like(pub),
    color="blue",
    alpha=0.4,
    s=70,
    label="공립일반고"
)

# 사립일반고 (빨간 점)
plt.scatter(
    pri,
    np.zeros_like(pri),
    color="red",
    alpha=0.4,
    s=70,
    label="사립일반고"
)

plt.yticks([])  # y축 제거
plt.xlabel("학교 주최 프로그램 키워드 개수")
plt.title("학교 유형별 학교 주최 프로그램 분포 (1차원 점 표현)")
plt.legend()
plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# numpy 배열로 변환
pub = np.array(public_counts)
pri = np.array(private_counts)

# 평균
pub_mean = np.mean(pub)
pri_mean = np.mean(pri)

# 분산 (표본분산: ddof=1)
pub_var = np.var(pub, ddof=1)
pri_var = np.var(pri, ddof=1)

# 표준편차
pub_std = np.std(pub, ddof=1)
pri_std = np.std(pri, ddof=1)

print("공립일반고")
print(f"평균: {pub_mean:.2f}")
print(f"분산: {pub_var:.2f}")
print(f"표준편차: {pub_std:.2f}")

print("\n사립일반고")
print(f"평균: {pri_mean:.2f}")
print(f"분산: {pri_var:.2f}")
print(f"표준편차: {pri_std:.2f}")
```
